```python
from sympy import *
from sympy.plotting import plot

x = Symbol('x')
fx = sin(x)**2 - cos(x)**2
```

```python
# 1. Определить корни
roots = solve(fx, x)
print(roots)
print(roots[0].evalf(), roots[1].evalf())
solveset(fx, x)
```

```python
# 4. Построить график
res = plot(fx)
```

```python
# 5. Вычислить вершину

diff_f = diff(fx, x)
print(diff_f)
print(solve(diff_f, x))
```

```python
# 2. Найти интервалы, на которых функция возрастает

solve(diff_f > 0, x)
```

```python
# 3. Найти интервалы, на которых функция убывает

solve(diff_f < 0, x)
```

```python
# 6. Определить промежутки, на котором f > 0

solve(fx > 0, x)
```

```python
# 7. Определить промежутки, на котором f < 0

solve(fx < 0, x)
```
